```python
import seaborn as sns

%matplotlib inline

# loading the built-in Tips dataset
tips = sns.load_dataset('tips')

tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

## Countplot just counts the number of occurences of a categorical column

```python
sns.countplot(x='sex', data=tips)
```

## Boxplot: Distribution of Quantitative data to show comparision between variables

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

## Violinplot

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

## Stripplot

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

## Swarmplot (in conjunction with Violinplot)

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```
